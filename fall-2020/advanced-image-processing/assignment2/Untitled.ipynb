```python
# setting the minimum and maximum thresholds  
# for double thresholding 
from sobel import sobel_edge_detection
from gaussian import gaussian_blur
import matplotlib.pyplot as plt
from PIL import Image

img = np.array(Image.open("images/test.jpg").convert('L'))
blur_img = gaussian_blur(img, kernel_size=1)

_, _, grad_mag, grad_dir = sobel_edge_detection(blur_img, convert_to_degree=True)
```

```python
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
plt.show()

plt.title('Blur Image')
plt.imshow(blur_img,cmap = 'gray')
plt.show()

plt.title('Sobel')
plt.imshow(grad_mag,cmap = 'gray')
plt.show()
```

```python
weak_th = None
strong_th = None

mag_max = np.max(grad_mag) 
if not weak_th:weak_th = mag_max * 0.1
if not strong_th:strong_th = mag_max * 0.5
```
